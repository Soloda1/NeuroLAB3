О задаче

Задача: Бинарная классификация.

Цель: На основе медицинских признаков пациента предсказать, есть ли у него диабет (Outcome = 1) или нет (Outcome = 0).

Целевая переменная:

Outcome (Диагноз диабета).

Признаки (Features):

Pregnancies (Беременности)

Glucose (Глюкоза)

BloodPressure (Артериальное давление)

SkinThickness (Толщина кожной складки)

Insulin (Инсулин)

BMI (Индекс массы тела)

DiabetesPedigreeFunction (Показатель наследственности по диабету)

Age (Возраст)

Для данных данных мы решили построить и оценить модель регрессии. В качестве модели взяли логистическую регрессию, так как задача классификации. 
Задача является бинарной классификацией (наличие или отсутствие диабета), для которой логистическая регрессия является оптимальным методом, так как она специально разработана для предсказания вероятности принадлежности к одному из двух классов.

Во-вторых, критически важным аспектом в медицинской диагностике является интерпретируемость модели. Логистическая регрессия предоставляет четкие коэффициенты, которые показывают влияние каждого медицинского признака на риск заболевания, что позволяет врачам понять логику принятия решений.

В-третьих, модель дает не просто ответ "болен/здоров", а вероятность от 0 до 100%. Это полезно, потому что врачи могут сами решать, при каком проценте риска стоит назначать дополнительные анализы. Например, если вероятность выше 70% — отправить на углубленное обследование.

В-четвертых, у нас не миллионы записей, а несколько сотен наблюдений. Логистическая регрессия хорошо работает на таких объемах, не перегружает систему и быстро обучается.

И последнее — результаты логистической регрессии часто совпадают с медицинской логикой. Например, она четко показывает, что уровень сахара в крови и лишний вес — это главные факторы риска. То есть модель не выдает что-то странное, а подтверждает уже известные врачам закономерности.
