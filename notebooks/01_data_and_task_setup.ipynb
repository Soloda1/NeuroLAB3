Задача регрессии (Regression).

1. 
Мы будем решать задачу прогнозирования стоимости недвижимости (price) на основе её характеристик. Это классическая задача регрессии, где целевая переменная (price) является непрерывной числовой величиной, а нам нужно построить модель, которая сможет предсказывать её значение по набору входных признаков.

2. 
Общее описание
Датасет содержит информацию о продажах жилой недвижимости в округе Кинг (King County, Washington, USA) за период 2014-2015 годов. Включает разнообразные характеристики домов, которые могут влиять на их рыночную стоимость.

Ключевые характеристики данных:
Количество записей: 918 (примерно)

Количество признаков: 20 (включая целевую переменную)

Период: май 2014 - май 2015

География: округ Кинг (Сиэтл и пригороды)

3. 
Целевая переменная:
price (float): Цена продажи дома (в долларах США). Диапазон значений: ~$80,000 - $3,000,000+.

Основные признаки:
Архитектурные характеристики:
bedrooms (int/float): Количество спален.

bathrooms (float): Количество ванных комнат (может быть дробным: 1.75, 2.5 и т.д.).

sqft_living (int): Жилая площадь в квадратных футах.

sqft_lot (int): Общая площадь участка в квадратных футах.

floors (float): Количество этажей (может быть дробным: 1.5, 2.5 и т.д.).

sqft_above (int): Площадь над землей (без подвала).

sqft_basement (int): Площадь подвала.

Качество и состояние:
grade (int): Оценка качества строительства и дизайна (1-13, где выше = лучше).

condition (int): Оценка общего состояния дома (1-5, где выше = лучше).

yr_built (int): Год постройки (1900-2015).

yr_renovated (int): Год последнего ремонта (0 = не ремонтировался).

Локационные признаки:
waterfront (int): Бинарный признак (0/1). Наличие вида на воду.

view (int): Качество вида от 0 до 4 (где выше = лучше вид).

zipcode (str): Почтовый индекс.

lat (float): Географическая широта.

long (float): Географическая долгота.

Контекстные признаки:
sqft_living15 (int): Средняя жилая площадь 15 ближайших соседей.

sqft_lot15 (int): Средняя площадь участка 15 ближайших соседей.

Вспомогательные признаки:
id (str): Уникальный идентификатор объекта.

date (str): Дата продажи в формате "YYYYMMDDThhmmss".

4.
Обоснование выбора задачи
Данная задача регрессии обладает значительной практической ценностью — разработанная модель может стать полезным инструментом для риелторов и инвесторов, помогая объективно оценивать стоимость недвижимости и принимать обоснованные финансовые решения.

С научной точки зрения датасет представляет интерес благодаря разнообразию признаков, включая географические координаты, что позволяет изучать пространственные закономерности и сравнивать различные подходы к регрессионному анализу.

Для образовательных целей этот классический датасет идеально подходит для освоения методов машинного обучения — от предобработки данных до интерпретации результатов, обеспечивая комплексное понимание реальных задач анализа данных.
